```python
!pip install matplotlib
import numpy as np

import matplotlib.pyplot as plt
```

```python
cofficent = np.array([[1],[-20],[25]],dtype= np.float32)
w = tf.Variable(0, dtype=tf.float32)
#tranning loop
optimizer = tf.optimizers.SGD(0.01)
for i in range(1000):
    with tf.GradientTape() as tape:
        cost = cofficients[0][0]*w**2 + cofficients[1][0]*w + cofficients[2][0]
    gradients = tape.gradient(cost, [w])
    optimizer.apply_gradients(zip(gradients, [w]))
print(f"optimal w: {w.numpy()}")
print(f"Minimum cost: {cost.numpy()}")
             
```

```python

# cost function
w_vals = np.linspace(-5,25,300)
cost_vals = w_vals**2 - 20*w_vals + 25

plt.figure(figsize=(8,5))
plt.plot(w_vals, cost_vals, color='steelblue',linewidth=2)
plt.xlabel('w')
plt.ylabel('cost(w)')
plt.title('cost curve')
plt.axvline(x=10, color="red", linestyle='--', label="Minimum at w=0")
plt.legend()
plt.grid(True, aplha=0.3)
plt.show()
            
```
