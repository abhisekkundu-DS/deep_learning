```python
# install pa kages
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
```

```python
texts = [
    'the cat sat on the mat',
    'the dog ran in the park',
    'cats and dog are pets',
    'i love deep learning'
]
```

```python
# Tokenise - map words to integers
tokenizer = Tokenizer(num_words = 10000,
                      oov_token = '<OOV>')
tokenizer.fit_on_texts(texts)

print('Vocabulary size : ', len(tokenizer.word_index))
print('word index', dict(list(tokenizer.word_index.items())[:10])) # print first 10 words in the vocabulary
```

```python
sequences = tokenizer.texts_to_sequences(texts)
print('Sequences : ', sequences[0])
```

```python
MAX_LEN = 10
padded = pad_sequences(sequences, maxlen= MAX_LEN, padding='post',truncating='post')
print('padded shape:',padded.shape)
```

```python
# The embedding layer
model = keras.Sequential(
[
    keras.layers.Embedding(input_dim=10000,
                           output_dim=64,
                           input_length=MAX_LEN),
    keras.layers.GlobalAveragePooling1D(),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])
model.summary()
```
